##**Feature Selection**

### **Regression Feature Selection**  
Numerical Input, Numerical Output

####Using K Best  
100 samples with 100 features are generated. 10 of them are informative (these are the ones that are used to generate the target values)
The best 10% features wanted to be selected. The Pearson's Correlation Coefficient (f_regression) is used.  
Regression  data is generated by `make_regression()`

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

# generate dataset
X, y = make_regression(n_samples=100, n_features=100, n_informative=10)
X.shape
```

```python
# define feature selection
fs = SelectKBest(score_func=f_regression, k=10) 
```

```python
# apply feature selection
X_selected = fs.fit_transform(X, y)
print(X_selected.shape)
```

#### Using Percentile  
100 samples with 120 features are generated. 12 of them are informative (these are the ones that are used to generate the target values)
The best 10% features wanted to be selected. The **Pearson's Correlation Coefficient** (`f_regression`) is used.

```python
from sklearn.datasets import make_regression
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_regression

# generate dataset
X, y = make_regression(n_samples=100, n_features=120, n_informative=12)
X.shape
```

```python
# define feature selection
fs = SelectPercentile(score_func=f_regression, percentile=10) 
```

```python
# apply feature selection
X_selected = fs.fit_transform(X, y)
print(X_selected.shape)
```

### **Classification Feature Selection**  
Numerical Input, Categorical Output

####Using K Best  
100 samples with 20 features are generated. Only 2 of them are informative (these are the ones that are used to generate the target values)
The best 2 features wanted to be selected. The **ANOVA F** measure (`f_classif`) is used.  
Classification data is generated by `make_classification()`

```python
# ANOVA feature selection for numeric input and categorical output
from sklearn.datasets import make_classification
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# generate dataset
X, y = make_classification(n_samples=100, n_features=20, n_informative=2)
```

```python
# define feature selection
fs = SelectKBest(score_func=f_classif, k=2)
```

```python
# apply feature selection
X_selected = fs.fit_transform(X, y)
print(X_selected.shape)
```
